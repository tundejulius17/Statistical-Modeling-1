```r
### 2
data<-read.table("makiwaraboard.txt", sep="\t", dec=".", header=TRUE)  # read data
attach(data)

interaction.plot(WoodType, BoardType, Deflection, FUN=mean)

model.main<-lm(Deflection~factor(WoodType)+factor(BoardType))
summary(model.main)

# a)

mu.32<-predict(model.main, newdata=data.frame(WoodType="Fir", BoardType="Tapered"), interval="confidence", level=0.95)

# fit     lwr      upr
# 55.70158 42.2666 69.13656
mu.32
```

```r
# b)
data.frame(data, fitted(model.main))

interaction.plot(WoodType, BoardType, fitted(model.main))
pred<-predict(model.main, newdata=expand.grid(BoardType=c("Stacked", "Tapered"), WoodType=c("Cherry", "Ash", "Fir", "Oak")))
pred
data.frame(expand.grid(BoardType=c("Stacked", "Tapered"), WoodType=c("Cherry", "Ash", "Fir", "Oak")), pred)
```

```r
# c)
model.interaction<-lm(Deflection~factor(WoodType)+factor(BoardType)+factor(WoodType):factor(BoardType))
summary(model.interaction)

betahat<-coef(model.interaction)

k1<-c(0,0,0,0,0,1,0,0)
k2<-c(0,0,0,0,0,0,1,0)
k3<-c(0,0,0,0,0,0,0,1)
K<-cbind(k1,k2,k3)

q<-3
Wald<-(t(t(K)%*%betahat)%*%solve(t(K)%*%vcov(model.interaction)%*%K)%*%t(K)%*%betahat)/q
Wald   #  0.1523881
p.value<-pf(Wald, 3, 328, lower.tail = FALSE)
p.value  # 0.9281153
```
